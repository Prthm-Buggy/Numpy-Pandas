```python
import pandas as pd
import numpy as np

data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma', 'Frank'],
    'age': [25, np.nan, 30, 22, np.nan, 22],  
    'salary': [50000, 60000, np.nan, 70000, 65000 ,43000],  
    'department': ['HR', np.nan, 'IT', 'Finance', 'HR', 'IT'],
}

df = pd.DataFrame(data)
print(df)
```

```python
from IPython.display import display
display(df)
```

```python
df.dropna(inplace=True)
df.dropna(subset=['age'], inplace=True)
df.dropna(axis=1, inplace=True)  # axis=1 means "columns"
display(df)
```
