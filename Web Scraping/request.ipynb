```python
import requests

url = 'https://youtube.com/watch?v=9bZkp7q19f0'  # Example URL, replace with the desired one
response = requests.get(url)
# Check if the request was successful
print("sucessful" if response.status_code == 200 else "connection failed")  # 200 means OK
```

```python
print(response.text)         # HTML content of the page
```
